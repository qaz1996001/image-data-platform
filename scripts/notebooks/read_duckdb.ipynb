```python
import pandas as pd
import duckdb
from openpyxl import load_workbook
```

```python
db_file = './medical_exams_streaming.duckdb'
with duckdb.connect(db_file,read_only = True) as con:
    print(type(con))
    result = con.execute('select * from medical_examinations_fact limit 10').df()
print(result)
```

```python
result
```

```python
source = '急診'
status = '終審報告'
description = '%Aorta%'
with duckdb.connect(db_file,read_only = True) as con:
    print(type(con))
    result = con.execute('SELECT * FROM medical_examinations_fact WHERE exam_source = ? AND exam_status = ? AND exam_description like ?   LIMIT 10',
                         [source, status, description]
                        ).df()
result
```

```python
with duckdb.connect(db_file,read_only = True) as con:
    print(type(con))
    result = con.execute('SELECT count(*) FROM medical_examinations_fact WHERE exam_source = ? AND exam_status = ? AND exam_description like ?',
                         [source, status, description]
                        ).df()
result
```

```python
with duckdb.connect(db_file,read_only = True) as con:
    print(type(con))
    df_result = con.execute('SELECT * FROM medical_examinations_fact WHERE exam_source = ? AND exam_status = ? AND exam_description like ? order by report_certification_datetime',
                         [source, status, description]
                        ).df()
df_result
```
